```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.distplot(a=tips['total_bill'], 
             bins=30,
             kde=True,
             hist=True,
             color='skyblue')
plt.title('Распределение общего счёта (total_bill)')
plt.xlabel('Сумма счёта ($)')
plt.ylabel('Плотность / Частота')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.pairplot(data=iris, hue='species', palette='husl', 
             plot_kws={'alpha': 0.7, 's': 40}, diag_kws={'linewidth': 1})
plt.suptitle('Парные зависимости в наборе Iris', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

titanic = sns.load_dataset('titanic')

numeric_cols = titanic.select_dtypes(include=[np.number]).columns
corr_matrix = titanic[numeric_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            linewidths=0.5, square=True, cbar_kws={'shrink': 0.8})
plt.title('Матрица корреляций признаков (Titanic)')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.kdeplot(data=titanic, x='fare', hue='class', 
            fill=True, common_norm=False, alpha=0.4, 
            palette='Set1', linewidth=2)
plt.title('Распределение стоимости билета по классам')
plt.xlabel('Стоимость билета ($)')
plt.ylabel('Плотность распределения')
plt.xlim(0, 300)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

titanic_clean = titanic.dropna(subset=['age', 'fare'])

sns.lmplot(data=titanic_clean, x='age', y='fare', 
           hue='sex', col='class', col_wrap=3,
           palette='Set2', height=4, aspect=1.1,
           scatter_kws={'alpha': 0.5, 's': 30},
           line_kws={'linewidth': 2})
plt.suptitle('Зависимость стоимости билета от возраста', y=1.02)
plt.tight_layout()
plt.show()
```
