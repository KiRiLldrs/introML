```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.show()
```

```python
t = np.linspace(0, 10, 51)

f = np.cos(t) + np.sin(t)

plt.plot(t, f, color='green', label='f(t)')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.xlim(0.5, 9.5)
plt.ylim(-2.5, 2.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 51)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(x, np.exp(x), color='blue')
ax[0].set_title('Экспонента: e^x')
ax[0].set_xlabel('x')
ax[0].set_ylabel('e^x')
ax[0].grid(True)

ax[1].plot(x[x > 0], np.log(x[x > 0]), color='green')
ax[1].set_title('Натуральный логарифм: ln(x)')
ax[1].set_xlabel('x')
ax[1].set_ylabel('ln(x)')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Name': ['Анна', 'Антон', 'Мария', 'Макс', 'Макс', 'Алекс', 
             'Антон', 'Мария', 'Макс', 'Алекс', 'Антон', 'Анна'],
    'Points': [145, 2524, 343, 4212, 5212, 6421, 745, 842, 524, 3224, 212, 1214]
}

df = pd.DataFrame(data)

grouped = df.groupby('Name')['Points'].agg(['min', 'max', 'mean'])

grouped.plot(kind='bar', figsize=(10, 6), rot=0)
plt.title('Статистика очков по игрокам')
plt.xlabel('Игрок')
plt.ylabel('Очки')
plt.legend(title='Статистика')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print(grouped)
```

```python
import matplotlib.pyplot as plt

labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [25, 35, 20, 20]
colors = ['red', 'yellow', 'orange', 'purple']
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(8, 6))
plt.pie(sizes, 
        explode=explode, 
        labels=labels, 
        colors=colors,
        autopct='%1.1f%%',
        shadow=True, 
        startangle=90)
plt.title('Доли фруктов')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

stores = ['Store A', 'Store B', 'Store C', 'Store D', 'Store E']
sales = [50, 80, 60, 70, 90]

plt.figure(figsize=(10, 6))
plt.bar(stores, sales, color='skyblue', edgecolor='navy')
plt.title('Продажи по магазинам')
plt.xlabel('Магазин')
plt.ylabel('Количество товаров')
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('creditcard.csv')

plt.figure(figsize=(12, 5))
plt.hist(df['Time'], bins=50, color='steelblue', edgecolor='black')
plt.title('Распределение транзакций по времени')
plt.xlabel('Время (секунды)')
plt.ylabel('Количество транзакций')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

class_counts = df['Class'].value_counts().sort_index()

plt.figure(figsize=(6, 5))
class_counts.plot(kind='bar', color=['green', 'red'], edgecolor='black')
plt.title('Количество наблюдений по классам')
plt.xlabel('Class (0 = легальная, 1 = мошенничество)')
plt.ylabel('Число транзакций')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 5))
class_counts.plot(kind='bar', color=['green', 'red'], edgecolor='black', log=True)
plt.title('Количество наблюдений (логарифмический масштаб)')
plt.xlabel('Class')
plt.ylabel('Число транзакций (log)')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3, which='both')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df[df['Class'] == 0]['V1'], 
         bins=20, alpha=0.5, density=True, 
         label='Class 0', color='gray', edgecolor='black')
plt.hist(df[df['Class'] == 1]['V1'], 
         bins=20, alpha=0.5, density=True, 
         label='Class 1', color='red', edgecolor='black')
plt.title('Распределение признака V1 по классам')
plt.xlabel('V1')
plt.ylabel('Плотность распределения')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

class_labels = ['Legitimate (0)', 'Fraud (1)']
class_sizes = df['Class'].value_counts().sort_index().values
colors_pie = ['green', 'red']
explode_pie = (0, 0.1)

plt.figure(figsize=(7, 7))
plt.pie(class_sizes, 
        labels=class_labels, 
        colors=colors_pie,
        explode=explode_pie,
        autopct='%1.3f%%', 
        shadow=True, 
        startangle=90)
plt.title('Доли транзакций по классам')
plt.axis('equal')
plt.tight_layout()
plt.show()

print("Распределение классов:")
print(df['Class'].value_counts())
print(f"\nДоля мошеннических транзакций: {df['Class'].mean()*100:.3f}%")
```
